```python
%matplotlib inline
import pandas as pd
```

## 1번문제

```python
df = pd.read_csv('data/movie_meta.csv')
df
```

```python
q1 = df['budget'].quantile(0.25)
q2 = df['budget'].quantile(0.75)
irq = q2-q1
irq
```

```python
cond = df['budget']>(q2 + 5*irq)
df[cond]
```

```python
df.drop(df[cond].index,inplace=True)
df
```

```python
df.plot(kind='scatter',x='budget',y='imdb_score')
```
