```python
df3 = pd.read_csv('data/beer.csv')
df3
```

```python
df3.plot(kind='box', y='abv')
```

```python
df3['abv'].describe()
```

```python
q1 = df3['abv'].quantile(0.25)
q3 = df3['abv'].quantile(0.75)
iqr = q3-q1
```

```python
cond = (df3['abv']<q1-1.5*iqr) | (df3['abv']>q3+1.5*iqr) # 이상점 조건
```

```python
df3[cond]
```

```python
df3.loc[2250,'abv']=0.055 #잘못된 값은 사용자가 수정
df3[cond]
```

```python
cond = (df3['abv']<q1-1.5*iqr) | (df3['abv']>q3+1.5*iqr) # 이상점 조건
df3[cond] # 아예 지워버려야 할 값
```

```python
df3[cond].index
```

```python
df3.drop(df3[cond].index,inplace=True)
df3
```

## 정확성 3 : 관계적 이상점(Relational Outlier)

```python
df4 = pd.read_csv('data/exam_outlier.csv')
df4
```

```python
df4.plot(kind='scatter',x='reading score',y='writing score')
```

```python
df4.corr() # 읽기, 쓰기 항목이 연관성 낮게 나옴(이상점 때문)
```

```python
df4[df4['writing score']>100]
```

```python
df4.drop(51,inplace=True)
```

```python
df4.plot(kind='scatter',x='reading score',y='writing score')
```

```python
df4.corr()
```

```python
cond = (df4['writing score']>90) & (df4['reading score']<40)
df4[cond]
```

```python
df4.drop(373,inplace=True)
df4.plot(kind='scatter',x='reading score',y='writing score')
```

```python
df4.corr()
#이상점을 제거하고 난 뒤 상관관계가 높아진 것을 확인할 수 있음
```
