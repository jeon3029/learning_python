```python
import pandas as pd
```

```python
df = pd.read_csv('data/laptops.csv')
df
```

```python
brand_nation={
    'Dell':'U.S',
    'Apple':'U.S',
    'Acer':'Taiwan',
    'HP':'U.S',
    'Lenovo':'China',
    'Alienware':'U.S',
    'Microsoft':'U.S',
    'Asus':'Taiwan'
}
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
```

```python
df
```

```python
nation_g = df.groupby('brand_nation')
type(nation_g)
```

```python
nation_g.count()
```

```python
nation_g.max()
```

```python
nation_g.mean()
```

```python
nation_g.first()
```

```python
nation_g.last()
```
