```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/titanic.csv',index_col=0)
df
```

```python
sns.violinplot(data=df,x='Age')
```

```python
df.sort_values(by='Fare',ascending=False)
```

```python
df['Survived'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
sns.violinplot(data=df,x='Pclass',y='Survived')
```

```python
ll = df[df['Survived']==1]
ll.plot(kind='hist',y='Age',x='Survived')
```

```python
df.loc[:,['Age','Survived']].corr()
```

## 정답

```python
# 나이 분포
df.plot(kind='hist', y='Age', bins=30)
```

```python
# 나이별 요금 분포
df.plot(kind='scatter', x='Age', y='Fare')
```

```python
df['Survived'].value_counts()
```

```python
# 객실 등급별 생존율 분포
sns.kdeplot(df['Pclass'], df['Survived'])
```

```python
# 생존 여부에 따른 나이 분포
sns.stripplot(data=df, x="Survived", y="Age")
```

```python
# 생존 여부에 따른 나이 및 성별 분포
sns.stripplot(data=df, x="Survived", y="Age", hue="Sex")
```
