```python
%matplotlib inline
import pandas as pd
```

```python
df  = pd.read_csv('data/silicon_detail.csv')
df
```

```python
ad = (df['company']=='Adobe') & (df['race'] == 'Overall_totals') 
ad2 = (df['count']!=0) & (df["job_category"] != "Totals") & (df["job_category"] != "Previous_totals")
ad = df[ad & ad2]
ad = ad.loc[:,['job_category','count']]
# ad.loc['job_category']
```

```python
ad.set_index('job_category',inplace=True)
ad
```

```python
ad.plot(kind='pie',y='count')
```
